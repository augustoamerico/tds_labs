```python
import spacy
```

```python
nlp = spacy.load("pt_core_news_sm")
```

```python
doc = nlp("A Apple está a ponderar comprar a startup do Reino Unido por 1 milhão de dolares")
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```

```python
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
def print_entities(doc):
    for ent in doc.ents:
        print(ent.text, ent.start_char, ent.end_char, ent.label_)

def print_tokens(doc):
    for token in doc:
        print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```

```python
doc_test = nlp("O Presidente declarou aberta a sessão às 15 horas e 2 minutos.")
print_entities(doc_test)
```
